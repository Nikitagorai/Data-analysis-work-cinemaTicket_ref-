```python
pivot1=\
dataset.pivot_table(index='tickets_sold',values='show_time',aggfunc=np.mean)
plt.figure(figsize=(30,25))
pivot1.plot(kind='bar',color='red')
plt.xlabel("tickets_sold")
plt.ylabel('show_time')
plt.show()
```

```python
pivot2=\
dataset.pivot_table(index='cinema_code',values='total_sales',aggfunc=np.mean)
plt.figure(figsize=(20,30))
pivot2.plot(kind='bar',color='green')
plt.xlabel("cinema_code")
plt.ylabel('total_sales')
plt.show()
```

```python
pivot3=\
dataset.pivot_table(index='ticket_use',values='day',aggfunc=np.mean)
plt.figure(figsize=(20,30))
pivot2.plot(kind='bar',color='yellow')
plt.xlabel("ticket_use")
plt.ylabel('day')
plt.show()
```

```python
pivot4=\
dataset.pivot_table(index='ticket_use',values='month',aggfunc=np.mean)
plt.figure(figsize=(20,30))
pivot2.plot(kind='bar',color='blue')
plt.xlabel("ticket_use")
plt.ylabel('month')
plt.show()
```
